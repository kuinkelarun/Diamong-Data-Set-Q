<h3> Attributes: </h3>

The attributes provide the framework in which the data are analyzed. These attributes in some cases are the limitations in which the given data can be interpreted. 

Attributes in diamonds data sets are the carat, cut, color, clarity, depth, table, price, x, y, and z. We may interpret the data based on these attributes. But if we were to ask a question, if the place of origin of the diamond impact the price of the diamond, we do not have basis to interpret this question. In that sense, attributes are the parameters in which we can interpret our data. 

Following are a brief explanation of the attributes in the "diamonds" data, the range of values are based on "diamonds" data: 
- *carat* - Weight of the diamonds, (with min 0.2 - max 5.01)

- > According to Tiffany & Co., "A diamond may have a higher carat weight without appearing larger and two diamonds of the same carat weight can vary in size if one is cut deeper than the other. In other words, it is important to note that carat weight does not necessarily denote size."

- *depth* - Top to bottom measurement of diamond, height from the cutlet to the table (with min 43.00 - max 79.00, measured usually in millimeters)

- *table* - The flat facet of the diamond, it is relative to the width of the diamond (with min 43.00 - max 95.00)

- *price* - Price of diamond per carat in US dollars (with min $326 - max $18,823)

It is important to note the following fact when it comes to the value of a diamond. 

- > The website tiffany.com says that "Taken by itself, carat weight does not determine a diamondâ€™s value. For example, two one-carat stones can vary widely in price when clarity, color and cut are taken into consideration."
- *color* - Our dataset has color ranging D - J. In general, diamond color ranges from D - Z, D-F being Colorless (best quality), G-J being Near Colorless (best value, as they look white like D-F when faced up), K-Z Faint Light Color (normally not sold in market, but they are available)

- *cut* - Cuts ranges from Fair, Good, Very Good, Premium, Ideal. Cut of the diamonds determine the quality of a diamond, as it determines how much light the stone reflects. For example, a D-color diamond with, high carat and clarity with not so good cut may result a diamond being dull

- *clarity* - It is a measurement of how clear the diamond is, which ranges from the best IF - worst l1.
    - IF - Internally Flawless
    - VVS1 & VVS2 - Very Slightly Included 
    - VS1 & VS2- Very Very Slightly Included
    - SI1 & SI2 - Slightly Included
    - I1 & I2 - Imperfect 
    - x - length in mm
    - y - width in mm 
    - z - depth in mm

Clarity being one among "four C's", a diamond industry term that directly impacts the price of a diamond.
- >According to Tiffany & Co., "A diamond with a poor clarity grade has multiple inclusions, which directly affects sparkle. Because inclusions hinder the refraction and return of light, the lower the clarity grade, the cloudier the diamond will appear."

<h4>Handling Zeroes: </h4>

The data that we have is fairly clean, however there are some missing values in the x, y and z attributes. Let us examine them, find out any inconsistency and handle them accordingly. 

```python
# locating the places where the value of x is zero
df_original = df

df.loc[df['x']==0]
```

We can see that most of the y and z also have zero values, where the values of x are zeroes. In that case we can either replace all the values with the median value of the respective attribute or we can simply remove those rows where x, y and z have zero values.

```python
# removind the row that contain y = 0, because if we 
# remove the rows with x = 0, we will also loose y, which has data value

df = df.loc[~(df['y'] == 0)]
```

```python
# checking again to make sure that x = 0 are removed. 
# And we can see that there is still one x = 0 in row 11182

df.loc[df['x']==0]
```

```python
# this shows that there are no more y = 0 values in our data set

df.loc[df['y']==0]
```

Using imputation method, we can impute the values of x and z with respective median values, as they are numerical data. 

```python
# replace x with median value
df.loc[11182, 'x'] = df['x'].median()
```

```python
# replace z with median value
df.loc[11182, 'z'] = df['z'].median()
```

```python
# again check to see if there are any zero values in x
df.loc[df['x']==0]
```
